```python
def read_orbit(filename):
    orbit_host = np.loadtxt('../../orbits/data/sag/' + filename + '_host.txt')
    orbit_sat = np.loadtxt('../../orbits/data/sag/' + filename + '_sat.txt')
    t_h = orbit_host[:,0]
    x_h = orbit_host[:,1]
    y_h = orbit_host[:,2]
    z_h = orbit_host[:,3]
    vx_h = orbit_host[:,4]
    vy_h = orbit_host[:,5]
    vz_h = orbit_host[:,6]
    
    t_s = orbit_sat[:,0]
    x_s = orbit_sat[:,1]
    y_s = orbit_sat[:,2]
    z_s = orbit_sat[:,3]
    vx_s = orbit_sat[:,4]
    vy_s = orbit_sat[:,5]
    vz_s = orbit_sat[:,6]
    
    R = np.sqrt((x_s-x_h)**2.0 + (y_s-y_h)**2.0+ (z_s-z_h)**2.0)
    V = np.sqrt((vx_s-vx_h)**2.0 + (vy_s-vy_h)**2.0+ (vz_s-vz_h)**2.0)
    
    return t_h, R, V
```

```python
t1, R1, V1 = read_orbit('sag_Herndf')
t2, R2, V2 = read_orbit('sag_Herndf2')
```

```python
t3, R3, V3 = read_orbit('sag_Herndf')
```

```python
plot(t1, R1, label='Besla 07, $alpha=0.45$')
plot(t3, R3, label= 'Besla07, $alpha=0.2$')
plot(t2, R2, label= 'VDM 13')
legend(fontsize=15)
xlabel('$Time [Gyrs]$', fontsize=25)
ylabel('$R [Kpc]$', fontsize=25)
```

```python
plot(t1, V1, label='Besla 07')
plot(t2, V2, label= 'VDM 13')
legend(fontsize=15)
xlabel('$Time [Gyrs]$', fontsize=25)
ylabel('$V [Km/s]$', fontsize=25)
```
